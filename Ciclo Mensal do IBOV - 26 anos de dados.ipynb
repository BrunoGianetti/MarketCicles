## Apresentação do dataframe contendo os dados

```python
import pandas_datareader as pdr
import datetime
import numpy as np
import pandas as pd

ibov = pdr.get_data_yahoo('^BVSP', start=datetime.datetime(1995,1,2), end=datetime.datetime(2021,2,28))

ibov
```

## Distribuição dos dados

```python
display(ibov.describe())
```

## Variação percentual dos preços

```python
pct_c = ((ibov.Close-ibov.Open)/ibov.Close)*100
pct_c
```

## Preparando os dados
***

### A ideia aqui é tabelar, pregão a pregão, cada mês isolado em 26 anos.

```python
jan26 = pd.DataFrame(pct_c)
jan26.head(23)
```
